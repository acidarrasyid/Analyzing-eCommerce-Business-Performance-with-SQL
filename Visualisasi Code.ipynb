```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
df= pd.read_csv('tugas2.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
plt.plot(df.year, df.new_customer, color ='red', linewidth=2.5, marker='o')
plt.title('New Customers Per Year', fontsize = 14, pad=20, weight='bold')
plt.xlabel('Year')
plt.ylabel('Total Customer')
plt.xticks(range(2016, 2019))
sns.despine()
plt.grid(color='darkgray', linestyle=':')
plt.tight_layout()
plt.savefig('new_customer.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
plt.plot(df.year, df.repeating_customer, color ='green', linewidth=2.5, marker='o')
plt.title('Repeat Customers Order Per Year', fontsize = 14, pad=20, weight='bold')
plt.xlabel('Year')
plt.ylabel('Total Customer')
plt.xticks(range(2016, 2019))
sns.despine()
plt.grid(color='darkgray', linestyle=':')
plt.tight_layout()
plt.savefig('repeating_customers.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
plt.plot(df.year, df.avg_mau, color ='black', linewidth=2.5, marker='o')
plt.title('MAU (Monthly Active Users) Per Year', fontsize = 14, pad=20, weight='bold')
plt.xlabel('Year')
plt.ylabel('Average Customer')
plt.xticks(range(2016, 2019))
sns.despine()
plt.grid(color='darkgray', linestyle=':')
plt.tight_layout()
plt.savefig('MAU.png')
plt.show()
```

```python
pd.set_option('display.max_columns', None)
df2 = pd.read_csv('tugas4.csv')
df2.head()
```
